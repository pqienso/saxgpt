```python
import torch
import torchaudio
from pathlib import Path
from src.data.util.tokenization import detokenize

def save_audio_from_codes(
    codes_path: Path,
    index: int,
    output_dir: Path = Path("outputs/")
):
    example = torch.load(codes_path, weights_only=False)
    lead = detokenize(example[index]["lead"])
    backing = detokenize(example[index]["backing"])

    print(example[index])
    
    torchaudio.save(output_dir / "inspect_lead.wav", lead, 32000)
    torchaudio.save(output_dir / "inspect_backing.wav", backing, 32000)
    
```
